
# 3. 布林通道


* 上下差距 < 10% 普通
* 上下差距 > 10% 開口

* 股價突破上軌 + 帶量 買
* 跌破中間值賣

````python
def plot_stock(df, buy_list, sell_list):
    df1 = df
    df1['date'] = pd.to_datetime(df1['date'])
    df1['date'] = df1['date'].apply(mpl_dates.date2num)
    df1 = df1.loc[:,['date', 'open', 'max', 'min', 'close']]
    fig, ax = plt.subplots(figsize=(16, 9))
    candlestick_ohlc(ax, df1.values, width=0.6, colorup='red', colordown='green', alpha=0.5)    
    date_format = mpl_dates.DateFormatter('%d %b %Y')
    ax.xaxis.set_major_formatter(date_format)
    for buy in buy_list:
        plt.plot(df1['date'][buy[0]], buy[1], 'b^', markersize='12')
    
    for sell in sell_list:
        plt.plot(df1['date'][sell[0]], sell[1], 'kv', markersize='12')
    fig.show()


TaiwanStockPriceDay['EMA5'] = talib.EMA(TaiwanStockPriceDay['close'], timeperiod=5)
TaiwanStockPriceDay['EMA20'] = talib.EMA(TaiwanStockPriceDay['close'], timeperiod=20)
TaiwanStockPriceDay['EMA30'] = talib.EMA(TaiwanStockPriceDay['close'], timeperiod=30)
TaiwanStockPriceDay['RSI'] = talib.RSI(TaiwanStockPriceDay['close'], timeperiod=14)
#TaiwanStockPriceDay['RSI12'] = talib.RSI(TaiwanStockPriceDay['close'], timeperiod=12)
#TaiwanStockPriceDay['RSI24'] = talib.RSI(TaiwanStockPriceDay['close'], timeperiod=24)
TaiwanStockPriceDay['TP'] = (TaiwanStockPriceDay['close'] + TaiwanStockPriceDay['min'] + TaiwanStockPriceDay['max']) / 3
TaiwanStockPriceDay['std'] = TaiwanStockPriceDay['TP'].rolling(20).std(ddof=0)
TaiwanStockPriceDay['MA-TP'] = TaiwanStockPriceDay['TP'].rolling(20).mean()
TaiwanStockPriceDay['BOLU'] = TaiwanStockPriceDay['MA-TP'] + 2*TaiwanStockPriceDay['std']
TaiwanStockPriceDay['BOLD'] = TaiwanStockPriceDay['MA-TP'] - 2*TaiwanStockPriceDay['std']
TaiwanStockPriceDay.head(68)
````

## 1. EMA Cross

```
if EMA5 > EMA20
    buy
elif EMA5 < EMA20
    sell
```

````python
# 記錄各時間採取的動作，和價位 (1962 * 2)
# 紀錄每筆買賣的狀況：投報率

trajectory_list = []
trading_list = []
buy_sell_list = []
flag = 0
temp = []
temp1 = []

for i in range(len(TaiwanStockPriceDay[34:])):
    if TaiwanStockPriceDay['EMA5'][i+32] < TaiwanStockPriceDay['EMA20'][i+32] and TaiwanStockPriceDay['EMA5'][i+33] > TaiwanStockPriceDay['EMA20'][i+33]:
        # buy at close
        trajectory_list.append(['buy', TaiwanStockPriceDay['close'][i+34]])
        flag = 1
        temp.append(TaiwanStockPriceDay['close'][i+34])
        temp1.append(i)
        
    elif TaiwanStockPriceDay['EMA5'][i+32] > TaiwanStockPriceDay['EMA20'][i+32] and TaiwanStockPriceDay['EMA5'][i+33] < TaiwanStockPriceDay['EMA20'][i+33]:
        trajectory_list.append(['sell', TaiwanStockPriceDay['close'][i+34]])
        if flag == 1:
            temp.append(TaiwanStockPriceDay['close'][i+34])
            temp1.append(i)
            temp1.append((temp[1]-temp[0])/temp[0])
            flag = 0
            trading_list.append(temp)
            buy_sell_list.append(temp1)
            temp = []
            temp1 = []
    else:
        trajectory_list.append(['hold', ''])
        
#pp.pprint(trajectory_list)
#pp.pprint(trading_list)

smaller_num = 0
medium_num = 0
bigger_num = 0
total_num = len(trading_list)
trade_sum = 0

for trade in trading_list:
    #pp.pprint((trade[1]-trade[0])/trade[0])
    
    trade_sum += (trade[1]-trade[0])/trade[0]
    
    if (trade[1]-trade[0])/trade[0] > 0:
        smaller_num += 1
    if (trade[1]-trade[0])/trade[0] > 0.03:
        medium_num += 1
    if (trade[1]-trade[0])/trade[0] > 0.05:
        bigger_num += 1

print("【未篩選前】")
print('賺錢的交易比例：', smaller_num/total_num)
print('>3%的交易比例：', medium_num/total_num)
print('>5%的交易比例：', bigger_num/total_num)
print('平均報酬率：', trade_sum/total_num)
print('交易次數：',  total_num)
print()

print("【篩選後：>3%】")
trade_sum = 0
total_num = 0
for trade in trading_list:
    if (trade[1]-trade[0])/trade[0] > 0.03:
        trade_sum += (trade[1]-trade[0])/trade[0]
        total_num += 1
print('平均報酬率：', trade_sum/total_num)
print('交易次數：',  total_num)
print()

print("【篩選後：>5%】")
trade_sum = 0
total_num = 0
for trade in trading_list:
    if (trade[1]-trade[0])/trade[0] > 0.05:
        trade_sum += (trade[1]-trade[0])/trade[0]
        total_num += 1
print('平均報酬率：', trade_sum/total_num)
print('交易次數：',  total_num)

buy_list = []
sell_list = []

for bs in buy_sell_list:
    if bs[2] > 0.05:
        buy_list.append([bs[0], trajectory_list[bs[0]][1]])
        sell_list.append([bs[1], trajectory_list[bs[1]][1]])

plot_stock(TaiwanStockPriceDay[34:], buy_list, sell_list)
````
